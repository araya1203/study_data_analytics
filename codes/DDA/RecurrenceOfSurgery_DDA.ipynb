### quest 2
- 직업군이 상위 2위에 속하는 항목 : value_count
- 체중 > 70, 신장 > 170, 흡연여부 없는 분
- 해당 갯수

```python
df_ROS_job = df_ROS_forpatient['직업'].value_counts().iloc[:2]
df_ROS_job # 직업군이 상위 2위에 속하는 항목 : value_count
```

```python
df_ROS_job.index
```

```python
# condition
condition = (df_ROS_forpatient['직업'] == df_ROS_job.index[0]) | (df_ROS_forpatient['직업'] == df_ROS_job.index[1])
condition2 = (df_ROS_forpatient['체중'] > 70) & (df_ROS_forpatient['신장'] > 170) & (df_ROS_forpatient['흡연여부'] == 0) 
```

```python
condition
```

```python
df_ROS_quest2 = df_ROS_forpatient.loc[condition & condition2]
```

```python
df_ROS_quest2
```

```python
df_ROS_quest2['직업'].value_counts()
```

```python
df_ROS_quest2['체중'].value_counts()
```
